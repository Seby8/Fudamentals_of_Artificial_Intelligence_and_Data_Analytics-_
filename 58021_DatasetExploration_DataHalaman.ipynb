## <b>Features (columns) of the dataset and what do they imply.</b>

- The features or column headers (Index) in the dataset represents the interaction of github user to the listed github repositories. Thinning down the dataset from the needed data provides clarity and could help the group to create an implication. There are 4 (four) interaction of github repositories on github; Star, Fork, Watch, Issue, and Pull request. 

```python
## Counting the github repository for each topic
CS4_topic_count = CS4_copy_df['topic'].value_counts()
CS4_topic_count
```

```python
CS4_features = CS4_copy_df[['topic','stars','forks','watches','issue','pull_requests']]
CS4_features
```

```python
CS4_features.dtypes
```

```python
CS4_features['stars'] = CS4_features['stars'].astype('float')
CS4_features['forks'] = CS4_features['forks'].astype('float')
CS4_features['watches'] = CS4_features['watches'].astype('float')
CS4_features['issue'] = CS4_features['issue'].astype('float')
CS4_features['pull_requests'] = CS4_features['pull_requests'].astype('float')
```

```python
CS4_features.dtypes
```

## Profile of the features

Below is the list of the available features  within the dataset. Each feature contains the number of interaction of github users to popular Github repositories 2020.

- <b>Stars</b> - This feature indicating that some people like a project, or people use them as bookmarks so they can follow what's going on with the repository later. 

- <b>Forks</b> - Here is the feature when people wants a copy of the repository/ they intend to make contributions to the repository.  

- <b>Watches</b> - This feature notified the github user who 'watch' the repository whenever there are any updates. 

- <b>Issue</b> - Here is a feature that lets the user to track where the development happens.

- <b>Pull request</b> - This feature let the user to tell other user about the changes on the project.
